```python
import pandas as pd

cv_data = pd.read_csv("cv_cleaned.csv")
jobs_data = pd.read_csv("final_jobs_dataset (1).csv")

cv_texts = cv_data['final_text']
job_texts = jobs_data['job_description']
```

```python
cv_texts = cv_texts.fillna("")
job_texts = job_texts.fillna("")
```

```python
print(cv_texts.isnull().sum())
print(job_texts.isnull().sum())
```

```python
cv_texts = cv_texts[:500]
job_texts = job_texts[:500]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(list(cv_texts) + list(job_texts))

cv_seq = tokenizer.texts_to_sequences(cv_texts)
job_seq = tokenizer.texts_to_sequences(job_texts)

max_len = 200

cv_pad = pad_sequences(cv_seq, maxlen=max_len, padding='post')
job_pad = pad_sequences(job_seq, maxlen=max_len, padding='post')
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, SimpleRNN, Dense

input_layer = Input(shape=(max_len,))
x = Embedding(10000, 128)(input_layer)
x = SimpleRNN(64)(x)
output = Dense(32, activation='relu')(x)

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
cv_embeddings = model.predict(cv_pad)
job_embeddings = model.predict(job_pad)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(cv_embeddings, job_embeddings)
```

```python
import numpy as np

for i in range(3):  # أول 3 CVs
    top_jobs = np.argsort(similarity_matrix[i])[-5:]

    print(f"\nCV {i} recommendations:")
    for j in top_jobs:
        print(jobs_data.iloc[j]['job_title'])
```

```python
for i in range(1):
    top_jobs = np.argsort(similarity_matrix[i])[-5:]

    print(f"\nCV {i} recommendations:")
    for j in top_jobs:
        print("Job:", jobs_data.iloc[j]['job_title'])
        print("Company:", jobs_data.iloc[j]['company'])
        print("Location:", jobs_data.iloc[j]['location'])
        print("-"*30)
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    return text

cv_texts_clean = [clean_text(t) for t in cv_texts]
job_texts_clean = [clean_text(t) for t in job_texts]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(cv_texts_clean + job_texts_clean)

cv_seq = tokenizer.texts_to_sequences(cv_texts_clean)
job_seq = tokenizer.texts_to_sequences(job_texts_clean)

max_len = 150  # قللناه شوية

cv_pad = pad_sequences(cv_seq, maxlen=max_len, padding='post')
job_pad = pad_sequences(job_seq, maxlen=max_len, padding='post')
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, SimpleRNN, Dense

input_layer = Input(shape=(max_len,))
x = Embedding(10000, 128)(input_layer)
x = SimpleRNN(64)(x)
output = Dense(32, activation='relu')(x)

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
from tensorflow.keras.layers import RepeatVector

input_layer = Input(shape=(max_len,))
x = Embedding(10000, 128)(input_layer)

# Encoder
x = SimpleRNN(64)(x)

# Decoder
x = RepeatVector(max_len)(x)
x = SimpleRNN(64, return_sequences=True)(x)
output = Dense(1)(x)

model = Model(inputs=input_layer, outputs=output)

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(cv_pad, cv_pad, epochs=3, batch_size=32)
```

```python
encoder = Model(inputs=model.input, outputs=model.layers[2].output)

cv_embeddings = encoder.predict(cv_pad)
job_embeddings = encoder.predict(job_pad)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(cv_embeddings, job_embeddings)
```

```python
import numpy as np

for i in range(3):
    top_jobs = np.argsort(similarity_matrix[i])[-5:]

    print(f"\nCV {i} recommendations:")
    for j in top_jobs:
        print(jobs_data.iloc[j]['job_title'])
```

# RNN Model Summary

In this section, an **RNN (Recurrent Neural Network)** model was built to analyze CV texts and job descriptions, then generate numerical representations (Embeddings) that help measure the similarity between candidates and job opportunities.

## Workflow of the Model

- The CV and job datasets were loaded and preprocessed.
- Missing values were handled and text cleaning was applied.
- A `Tokenizer` was used to convert words into numerical sequences.
- `Padding` was applied to make all text sequences the same length.

The RNN architecture consisted of:

- `Embedding Layer` to transform words into dense vector representations.
- `SimpleRNN Layer` to capture sequential patterns and contextual information from text.
- `Dense Layer` to generate the final embedding representation.

After training:
- Embeddings were extracted for both CVs and job descriptions.
- `Cosine Similarity` was used to calculate similarity scores between them.
- The system recommended the most relevant jobs based on the highest similarity scores.

---

# Autoencoder RNN Enhancement

The model was further improved using an **Encoder–Decoder (Autoencoder) Architecture**.

## Model Idea

- The **Encoder** compresses the text into a compact feature representation.
- The **Decoder** attempts to reconstruct the original sequence.
- This approach helps the model learn deeper semantic features from the text data.

The enhanced architecture included:
- `RepeatVector`
- `SimpleRNN` with `return_sequences=True`

The embeddings extracted from the Encoder were then used again for similarity calculation and recommendation generation.

---

# Results and Observations

- The model successfully recommended jobs related to:
  - Data Science
  - Data Engineering
  - Data Analysis
  - Machine Learning

- Similarity scores were extremely high (around 0.999), indicating that the model effectively captured relationships between similar CVs and job descriptions.

- After applying the Autoencoder architecture, the recommendations became more accurate and semantically relevant compared to the baseline RNN model.

The RNN-based recommendation system demonstrated the ability to understand and analyze textual information from CVs and job descriptions effectively.

Using the Encoder–Decoder architecture improved feature extraction quality and produced more relevant job recommendations based on user skills and experience.
